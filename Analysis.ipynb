```python
#Import necessary Python libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read the data csv and store it in a dataframe called df
df = pd.read_csv('2020_IN_Region_Mobility_Report.csv', sep=',')
```

```python
#inspect dataframe size and columns
print(df.shape, df.columns)
```

```python
#Check sub_regions for Delhi
print(df.sub_region_1.unique())
```

```python
#Save Delhi data in a separate dataframe
delhi = df[df['sub_region_1'] == 'Delhi']
print(delhi.describe())
```

```python
#Select the required columns and check if there are any null values in the dataframe
columns = ['date','retail_and_recreation_percent_change_from_baseline','grocery_and_pharmacy_percent_change_from_baseline',
       'parks_percent_change_from_baseline','transit_stations_percent_change_from_baseline','workplaces_percent_change_from_baseline']
delhi = delhi[columns]
print(delhi.isna().nunique())
```

```python
#Extract month number from the date column and add it to a new column
delhi['date_str'] = delhi.date.astype(str)
delhi['month'] = delhi['date_str'].str.slice(start = 5, stop = 7)
delhi_test = delhi.groupby('month').sum()
```

```python
#Create a lineplot to visualize monthly mobility trends
plt.figure(figsize=(12,8))
plt.ylabel('monthly aggregated anomaly')  
plt.xlabel('month number in 2020')
sns.lineplot(data=delhi_test)
```

```python
#Add the monthly mean NO2 concentrations as a list from Sentinel5P_Delhi_2020.ipynb (February onwards)
#Scale - x micromoles/sq. meter (where x = list element)
Avg_Monthly_No2_Delhi = [94.79486394733583, 74.83880458576214, 37.325032650345493, 60.12819499559, 62.37332805184553, 57.75366189501766, 43.184628384035926, 57.78902743020165, 80.93005255020477, 93.34697076631547, 133.59854525545276]
```

```python
#Aggregate all mobility variables into one statistic
delhi_test['overall_activity_anomaly'] = delhi_test.sum(axis=1)
print(delhi_test)
```

```python
Delhi_processed_data = pd.DataFrame()
```

```python
#For a better scale, divide the mobility value columns by 1000
Delhi_processed_data['Aggregate_overall_mobility'] = (delhi_test['overall_activity_anomaly'].values)/1000
print(Delhi_processed_data)
```

```python
Delhi_processed_data['Average_monthly_NO2_concentrations'] = Avg_Monthly_No2_Delhi
print(Delhi_processed_data)
```

```python
sns.lineplot(data=Delhi_processed_data)
```
